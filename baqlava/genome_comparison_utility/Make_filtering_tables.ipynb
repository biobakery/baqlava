```python
import pandas as pd
```

```python
# code developed in /n/holystore01/LABS/huttenhower_lab/Users/jjensen/baqlava_V1_sandbox/20260520_plasmid_investigation/
filesloc = "/n/holystore01/LABS/huttenhower_lab/Users/jjensen/baqlava_V1_sandbox/20260225_plasmid/baqlava/baqlava/utility_files/"

baqv1_markers = pd.read_csv(filesloc + 'idmap_nucleotide_nofilter.txt', sep="\t", names=['name1','name2','length'])
baqv1_proteins = pd.read_csv(filesloc + 'idmap_protein_nofilter.txt', sep="\t", names=['name1','name2','length'])

default_markers = pd.read_csv(filesloc + 'idmap_nucleotide_default.txt', sep="\t", names=['name1','name2','length'])
default_proteins = pd.read_csv(filesloc + 'idmap_protein_default.txt', sep="\t", names=['name1','name2','length'])

conservative_markers = pd.read_csv(filesloc + 'idmap_nucleotide_conservative.txt', sep="\t", names=['name1','name2','length'])
conservative_proteins = pd.read_csv(filesloc + 'idmap_protein_conservative.txt', sep="\t", names=['name1','name2','length'])

baqv1_VGBsdropped_MGX = pd.read_csv(filesloc + 'MGX_plasmid_VGBs.txt', sep="\t")
baqv1_VGBsdropped_MTX = pd.read_csv(filesloc + 'MTX_plasmid_VGBs.txt', sep="\t")

baqref = pd.read_csv("BAQLaVa_metadata_genomes.txt", sep="\t")[['Genome Name','Segment_Group']]
```

```python
# start by getting genome names for markers/proteins

def split_mapname(df, lent):
    df1 = df.copy()
    df1 = df1[df1['length']>= lent]
    name = []
    for i in df1['name1']:
        name.append(i.split("_")[0])
    df1['Genome Name'] = name
    return df1
    
baqv1_markers = split_mapname(baqv1_markers, 500)
baqv1_proteins = split_mapname(baqv1_proteins, 200)

default_markers = split_mapname(default_markers, 500)
default_proteins = split_mapname(default_proteins, 200)

conservative_markers = split_mapname(conservative_markers, 500)
conservative_proteins = split_mapname(conservative_proteins, 200)
```

```python
# merging all info into one table (markers):

t1 = pd.merge(baqref.copy(), baqv1_markers, on='Genome Name', how='outer').rename(columns={'name1':'Marker','length':'Marker_Length'}).drop(columns=['name2'])
t1['Filter Level: None'] = t1['Marker'].notna().astype(int)
t1 = t1.dropna(subset=['Marker'])

t2 = pd.merge(t1.copy(), default_markers, left_on=['Genome Name','Marker'], right_on=['Genome Name','name1'], how='left')
t2['Filter Level: Default'] = t2['name1'].notna().astype(int)
t2 = t2.drop(columns=['name1','name2','length'])

t3 = pd.merge(t2.copy(), conservative_markers, left_on=['Genome Name','Marker'], right_on=['Genome Name','name1'], how='left')
t3['Filter Level: Conservative'] = t3['name1'].notna().astype(int)
t3 = t3.drop(columns=['name1','name2','length'])

t4 = pd.merge(t3.copy(), baqv1_VGBsdropped_MGX.copy(), left_on='Segment_Group', right_on='feature', how='left')
t4['Filter Level: V1_full_VGBs'] = t4['feature'].isna().astype(int)
t4 = t4.drop(columns=['feature'])

#t4.to_csv("marker_plasmid_filtering_MGX.txt", sep="\t", index=False)
```
